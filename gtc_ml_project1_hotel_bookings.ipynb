```python
if "country" in clean_df.columns:
    clean_df["country"] = clean_df["country"].fillna(clean_df["country"].mode()[0])
```

```python
before = len(clean_df)
clean_df = clean_df.drop_duplicates()
print(f"✅ duplicates removed: {before - len(clean_df)} | Shape now: {clean_df.shape}")
```

```python
# تحويل reservation_status_date إلى datetime (لو موجود)
if "reservation_status_date" in clean_df.columns:
    clean_df["reservation_status_date"] = pd.to_datetime(clean_df["reservation_status_date"], errors="coerce")

# بناء arrival_date من (year + month النصي + day)
def _to_month_num(m):
    try:
        return pd.to_datetime(str(m), format="%B").month   # January
    except Exception:
        try:
            return pd.to_datetime(str(m), format="%b").month  # Jan
        except Exception:
            return np.nan

if all(c in clean_df.columns for c in ["arrival_date_year","arrival_date_month","arrival_date_day_of_month"]):
    mnum = clean_df["arrival_date_month"].apply(_to_month_num)
    clean_df["arrival_date"] = pd.to_datetime(
        dict(year=clean_df["arrival_date_year"], month=mnum, day=clean_df["arrival_date_day_of_month"]),
        errors="coerce"
    )
    # نشيل الشهر النصّي بعد ما بقى عندنا تاريخ موحَّد
    clean_df = clean_df.drop(columns=["arrival_date_month"])
    print("✅ built arrival_date & dropped arrival_date_month")
else:
    print("⚠️ أعمدة بناء arrival_date مش كاملة؛ متجاوزين الخطوة.")
```

```python

for c in ["adults","children","babies","stays_in_weekend_nights","stays_in_week_nights"]:
    if c not in clean_df.columns:
        clean_df[c] = 0

clean_df["total_guests"] = clean_df["adults"].fillna(0) + clean_df["children"].fillna(0) + clean_df["babies"].fillna(0)
clean_df["total_nights"] = clean_df["stays_in_weekend_nights"].fillna(0) + clean_df["stays_in_week_nights"].fillna(0)
clean_df["is_family"]   = ((clean_df.get("children",0) > 0) | (clean_df.get("babies",0) > 0)).astype(int)

print("✅ features added: total_guests, total_nights, is_family")
clean_df[["total_guests","total_nights","is_family"]].head()
```

```python
# E-1) إزالة أعمدة الـ leakage + تفكيك arrival_date لأعمدة رقمية
# leakage: لا نستخدمهم في التدريب عشان ما يسرّبوش النتيجة
for col in ["reservation_status", "reservation_status_date"]:
    if col in clean_df.columns:
        clean_df = clean_df.drop(columns=[col])

# تفكيك arrival_date لو موجودة
if "arrival_date" in clean_df.columns:
    ad = pd.to_datetime(clean_df["arrival_date"], errors="coerce")
    clean_df["arrival_year"]    = ad.dt.year
    clean_df["arrival_month"]   = ad.dt.month
    clean_df["arrival_day"]     = ad.dt.day
    clean_df["arrival_dow"]     = ad.dt.dayofweek  # Monday=0
    clean_df["arrival_quarter"] = ad.dt.quarter
    clean_df = clean_df.drop(columns=["arrival_date"])

print("✅ Dropped leakage (if present) + expanded arrival_date (if present)")
```

```python
# E-2) تشطيب الأنواع + ملء أي NaN رقمي متبقّي + حفظ
# - bool -> int
# - object -> category codes
# - fill numeric NaN with median
for c in clean_df.columns:
    s = clean_df[c]
    if pd.api.types.is_bool_dtype(s):
        clean_df[c] = s.astype(int)
    elif s.dtype == "object":
        clean_df[c] = s.astype("category").cat.codes  # -1 للـ NaN/غير معروف

num_cols = clean_df.select_dtypes(include=["number"]).columns
for c in num_cols:
    if clean_df[c].isna().any():
        clean_df[c] = clean_df[c].fillna(clean_df[c].median())

out_path = "hotel_bookings_after_E.csv"
clean_df.to_csv(out_path, index=False)
print(f"💾 Saved: {out_path} | Shape: {clean_df.shape}")
print("Top missing (should be zeros or near-zero):")
print(clean_df.isna().sum().sort_values(ascending=False).head(10))
```

```python
# F-1) تحديد الهدف وتقسيم 80/20 (seed=42)
target = "is_canceled"
if target not in clean_df.columns:
    raise ValueError("عمود is_canceled غير موجود.")

y = clean_df[target].astype(int)
X = clean_df.drop(columns=[target])

rng = np.random.default_rng(42)
idx = np.arange(len(X))
rng.shuffle(idx)

test_size = int(0.2 * len(X))
test_idx  = idx[:test_size]
train_idx = idx[test_size:]

X_train = X.iloc[train_idx].reset_index(drop=True)
X_test  = X.iloc[test_idx].reset_index(drop=True)
y_train = y.iloc[train_idx].reset_index(drop=True)
y_test  = y.iloc[test_idx].reset_index(drop=True)

print("Shapes →",
      "X_train:", X_train.shape,
      "| X_test:", X_test.shape,
      "| y_train:", y_train.shape,
      "| y_test:", y_test.shape)

print("Cancel rate (overall):", round(y.mean()*100, 2), "%")
```

```python
# F-2) حفظ ملفات الـ split
X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("💾 Saved: X_train.csv, X_test.csv, y_train.csv, y_test.csv")
```
